# 1) Importing the dataset

```python
import pandas as pd
dataset = pd.read_csv(r'C:\Users\user\Desktop\pytorch\py\iris.csv').values
print(dataset)
```

# 2) Devide into data & target 

```python
data=dataset[:,0:4]
target=dataset[:,4]
print(data)
print(target)

```

# 3) convert to categorical form

```python
from keras.utils import np_utils
new_target = np_utils.to_categorical(target)
print(new_target[60:70])
```

# Dataset split to train data set & test data set

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,new_target,test_size=0.1)
```

```python
print(train_data,train_data.shape)
```

```python
print(test_data,test_data.shape)
```

# Defining Neural Network Architecture 

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()   # empty sequential type neural network

model.add(Dense(128,input_dim=4,activation='relu'))  # 1st Hidden Layer
model.add(Dense(64,activation='relu')) #2nd Hidden Layer
model.add(Dense(3,activation='softmax')) #output layer

model.compile(loss='categorical_crossentropy',optimizer='adaGrad',metrics=['accuracy'])
model.summary()
```

# Initial weight & bias values 


```python
for layer in model.layers:
    
    print('================================',layer.name,'================================')
    
    parameters=layer.get_weights()
    print(parameters)
```

# Model Training 

```python
# training the model
model.fit(train_data,train_target,epochs=500,validation_split=0.1)
```

```python
from matplotlib import pyplot as plt

plt.plot(model.history.history['loss'],label='loss')
plt.plot(model.history.history['val_loss'],label='val loss')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(model.history.history['accuracy'],label='accuracy')
plt.plot(model.history.history['val_accuracy'],label='val accuracy')
plt.legend()
plt.xlabel('#epoch')
plt.xlabel('loss')
plt.show()
```

```python
import numpy as np

test_data_1=np.array([5.3,3.5,1.7,0.3]).reshape(1,-1)

predicted_target_1=model.predict(test_data_1)
print(predicted_target_1)
```

```python
label=np.argmax(predicted_target_1)
print(label)
```
